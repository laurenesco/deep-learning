## Optimization Overview

Now we need to understand how we can minimize our loss functions. This is called *training* the model.

#### Training a Model

Training a model consists of minimizing the models loss function. Formally, find:

$$ \theta^* = arg\;\underset{\theta}{min} \; L(\theta \mid\mathcal{D}) $$

So conceptually, the optimizer finds a parameter setting that minimizes the loss. Thus, it interacts with the loss function and the model, since it will be updating the parameters to fit the data best.

#### Gradient Descent

Gradient descent (or some variant of it) is what is behind optimization in deep networks. See the lecture from Machine Learning to see how it works.

How is gradient descent initialized? In practice? We start from a random location. We will initialized our parameters randomly according to a normal or uniform distribution.

Note that step size ($\epsilon$) plays a major role in gradient descent. Too large and you may oscillate or overshoot. Too small and your training may take too longor not reach minimia. Getting stuck in local minima is always a possibility (but usually not that big of a deal if the model is initialized properly. See lecture from machine learning for more details.

```python
import matplotlib.pyplot as plt
import torch

def f(x: torch.Tensor) -> torch.Tensor:
    return 3*x**2+torch.sin(5*x)

def grad_f(x: torch.Tensor) -> torch.Tensor:
    return 6*x+5*torch.cos(5*x)

plt.plot(torch.arange(-2,2,0.01), f(torch.arange(-2,2,0.01)))
```

```python
x = torch.randn(1)
epsilon = 0.1
print(x)

all_x = []

for _ in range(100):
    x = x - epsilon * grad_f(x)
    all_x.append(x)

x
plt.plot(torch.arange(-2,2,0.01), f(torch.arange(-2,2,0.01)))
plt.plot(torch.stack(all_x), f(torch.stack(all_x)), 'ro')
```
